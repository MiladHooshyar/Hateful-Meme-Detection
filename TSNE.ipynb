```python
import numpy as np
import matplotlib.pylab as plt
from sklearn.manifold import TSNE
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, confusion_matrix
import seaborn as sns
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(131)

TSNE_img = np.load('D:/meme_data/features/TSNE_img.npz')
Hateful = TSNE_img['Hatefull']
Non_hateful = TSNE_img['Non_hateful']

plt.title('Image features')
plt.scatter(Non_hateful[:, 0], Non_hateful[:, 1], alpha = 0.1, c = 'r',
            s = 50, edgecolors=None, label = 'Non-hateful')
plt.scatter(Hateful[:, 0], Hateful[:, 1], alpha = 0.1, c = 'k',
            s = 50, edgecolors=None, label = 'Hateful')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
##############################################
plt.subplot(132)
TSNE_img = np.load('D:/meme_data/features/TSNE_txt.npz')
Hateful = TSNE_img['Hatefull']
Non_hateful = TSNE_img['Non_hateful']

plt.title('Text features')
plt.scatter(Non_hateful[:, 0], Non_hateful[:, 1], alpha = 0.1, c = 'r',
            s = 50, edgecolors=None, label = 'Non-hateful')
plt.scatter(Hateful[:, 0], Hateful[:, 1], alpha = 0.1, c = 'k',
            s = 50, edgecolors=None, label = 'Hateful')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
################################
plt.subplot(133)
TSNE_img = np.load('D:/meme_data/features/TSNE_img_txt.npz')
Hateful = TSNE_img['Hatefull']
Non_hateful = TSNE_img['Non_hateful']

plt.title('Image+Text features')
plt.scatter(Non_hateful[:, 0], Non_hateful[:, 1], alpha = 0.1, c = 'r',
            s = 50, edgecolors=None, label = 'Non-hateful')
plt.scatter(Hateful[:, 0], Hateful[:, 1], alpha = 0.1, c = 'k',
            s = 50, edgecolors=None, label = 'Hateful')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
