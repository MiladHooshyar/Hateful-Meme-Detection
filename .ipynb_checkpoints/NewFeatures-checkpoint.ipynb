```python
import numpy as np
import matplotlib.pylab as plt
from sklearn.manifold import TSNE
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, confusion_matrix
import seaborn as sns
```

```python
train = np.load('D:/meme_data/features/train.npz')

txt_fea_train = train['txt_fea']
txt_mod_train = train['txt_mod']
img_fea_train = train['img_fea']
img_mod_train = train['img_mod']
label_train = train['label']


dev = np.load('D:/meme_data/features/dev.npz')
txt_fea_dev = dev['txt_fea']
txt_mod_dev = dev['txt_mod']
img_fea_dev = dev['img_fea']
img_mod_dev = dev['img_mod']
label_dev = dev['label']
```

# Baseline: Logistic Regression

```python
X_train = np.concatenate((img_mod_train, txt_mod_train, txt_fea_train), axis=1)
y_train = np.copy(label_train)

X_dev = np.concatenate((img_mod_dev, txt_mod_dev, txt_fea_dev), axis=1)
y_dev = np.copy(label_dev)


clf = LogisticRegression(random_state=0).fit(X_train, y_train)
prob = clf.predict_proba(X_train)
print('Training AUC ROC with image:', roc_auc_score(y_train, prob[:,1]))

prob = clf.predict_proba(X_dev)
print('Validation AUC ROC with image:', roc_auc_score(y_dev, prob[:,1]))

print('******************')

confusion = confusion_matrix(
y_dev, prob.argmax(axis = 1))
confusion_normalized = confusion.astype("float") / confusion.sum(axis=1)

axis_labels = ['Non-hateful', 'Hateful']
ax = sns.heatmap(
    confusion_normalized, xticklabels=axis_labels, yticklabels=axis_labels,
    cmap='Blues', annot=True, fmt='.2f', square=True)
plt.title("Confusion matrix")
plt.ylabel("True label")
plt.xlabel("Predicted label")
plt.show()
```

# New model

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import PIL
import tensorflow as tf
import json

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
import seaborn as sns
from tensorflow.keras.models import Model

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, confusion_matrix

from sklearn.model_selection import train_test_split
from tensorflow.keras.regularizers import l2

from sklearn import metrics
from tensorflow.keras import backend as K

from tensorflow.keras import regularizers
```

```python
train = np.load('D:/meme_data/features/train.npz')

txt_fea_train = train['txt_fea']
txt_mod_train = train['txt_mod']
img_fea_train = train['img_fea']
img_mod_train = train['img_mod']
label_train = train['label']


dev = np.load('D:/meme_data/features/dev.npz')
txt_fea_dev = dev['txt_fea']
txt_mod_dev = dev['txt_mod']
img_fea_dev = dev['img_fea']
img_mod_dev = dev['img_mod']
label_dev = dev['label']


test = np.load('D:/meme_data/features/test.npz')
txt_fea_test = test['txt_fea']
txt_mod_test = test['txt_mod']
img_fea_test = test['img_fea']
img_mod_test = test['img_mod']
id_test = np.load('D:/meme_data/features/test_id.npy')

num_classes = len(np.unique(label_train))
```

```python
X = np.concatenate((txt_fea_train, txt_mod_train, img_fea_train, img_mod_train), axis=1)
y = np.copy(label_train)
X = (X - X.mean(axis = 0)) / X.std(axis = 0)

X_dev = np.concatenate((txt_fea_dev, txt_mod_dev, img_fea_dev, img_mod_dev), axis=1)
y_dev = np.copy(label_dev)
X_dev = (X_dev - X.mean(axis = 0)) / X.std(axis = 0)

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, shuffle= True)
```

```python
img_input = layers.Input(shape=(img_fea_train.shape[1],))
x = layers.Dense(50,kernel_regularizer=regularizers.l2(0.03))(img_input)
x = layers.Dense(20, activation='relu',kernel_regularizer=regularizers.l2(0.01))(x)
x = layers.Dropout(0.5)(x)
x = layers.Dense(10, activation='relu')(x)
x = layers.Dropout(0.5)(x)
img_out = layers.Dense(5, activation='relu')(x)

txt_input = layers.Input(shape=(txt_fea_train.shape[1],))
x = layers.Dense(50,kernel_regularizer=regularizers.l2(0.03))(txt_input)
x = layers.Dense(20, activation='relu',kernel_regularizer=regularizers.l2(0.01))(x)
x = layers.Dropout(0.5)(x)
x = layers.Dense(10, activation='relu')(x)
x = layers.Dropout(0.5)(x)
txt_out = layers.Dense(5, activation='relu')(x)

txt_mod_input = layers.Input(shape=(txt_mod_train.shape[1],))
x = layers.Dense(50, kernel_regularizer=regularizers.l2(0.03))(txt_mod_input)
x = layers.Dense(20, activation='relu', kernel_regularizer=regularizers.l2(0.01))(x)
x = layers.Dropout(0.5)(x)
x = layers.Dense(10, activation='relu')(x)
x = layers.Dropout(0.5)(x)
txt_mod_out = layers.Dense(100, activation='relu')(x)

img_mod_input = layers.Input(shape=(img_mod_train.shape[1],))
img_mod_out = layers.Dense(5, activation='sigmoid')(img_mod_input)
# x = layers.Dense(64, activation='relu')(x)
# img_mod_out = layers.Dense(5, activation='relu')(x)


concatenated = layers.concatenate([img_out, txt_out, txt_mod_out, img_mod_out])

# concatenated = layers.concatenate([img_out, txt_out])


classifier = layers.Dense(50, activation='sigmoid')(concatenated)
x = layers.Dropout(0.5)(classifier)
x = layers.Dense(10, activation='relu')(x)
x = layers.Dropout(0.5)(x)
classifier = layers.Dense(num_classes)(classifier)

model = Model([img_input, txt_input, txt_mod_input, img_mod_input], classifier)
# model = Model([img_input, txt_input], classifier)


model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
#               loss=tf.keras.losses.MeanSquaredError(),
              metrics=['accuracy'])

model.summary()
```

```python
wei = np.where(label_train==1, 3., 1)


epochs = 50
history = model.fit(
    [img_fea_train, txt_fea_train, txt_mod_train, img_mod_train], label_train,
#     [img_fea_train, txt_fea_train], label_train,
    
    epochs=epochs, 
#     validation_split=0.2, 
    validation_data= ([img_fea_dev, txt_fea_dev, txt_mod_dev, img_mod_dev], label_dev),
    sample_weight=wei,
#     validation_data= ([img_fea_dev, txt_fea_dev], label_dev),
    batch_size = 1000
)
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss=history.history['loss']
val_loss=history.history['val_loss']

epochs_range = range(epochs)

plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

```python
pred = model.predict([img_fea_dev, txt_fea_dev, txt_mod_dev, img_mod_dev])
prob = np.exp(pred) / np.exp(pred).sum(axis = 1)[:,None]
print('Validation AUC ROC with image:', roc_auc_score(label_dev, prob[: , 1], multi_class='ovo'))
```

```python
import pandas as pd
from collections import defaultdict

pred = model.predict([img_fea_test, txt_fea_test, txt_mod_test, img_mod_test])
prob = np.exp(pred) / np.exp(pred).sum(axis = 1)[:,None]
proba = prob[:, 1]
lab = np.where(proba> 0.5, 1, 0)

res = defaultdict()
res['id'] = list(id_test)
res['proba'] = list(proba)
res['label'] = list(lab)

pd.DataFrame.from_dict(res).to_csv('res_1.csv', index = False)
```

```python
train = np.load('D:/meme_data/features/train.npz')

txt_fea_train = train['txt_fea']
txt_mod_train = train['txt_mod']
img_fea_train = train['img_fea']
img_mod_train = train['img_mod']
label_train = train['label']


dev = np.load('D:/meme_data/features/dev.npz')
txt_fea_dev = dev['txt_fea']
txt_mod_dev = dev['txt_mod']
img_fea_dev = dev['img_fea']
img_mod_dev = dev['img_mod']
label_dev = dev['label']


test = np.load('D:/meme_data/features/test.npz')
txt_fea_test = test['txt_fea']
txt_mod_test = test['txt_mod']
img_fea_test = test['img_fea']
img_mod_test = test['img_mod']
id_test = np.load('D:/meme_data/features/test_id.npy')

num_classes = len(np.unique(label_train))
```

```python
N_feature = X_train.shape[1]

model = Sequential([
    layers.Dense(500, input_dim=N_feature, activation='sigmoid'),
    layers.Dense(5000, activation= tf.keras.layers.LeakyReLU(alpha=0.1)),
    layers.Dense(100, activation= tf.keras.layers.LeakyReLU(alpha=0.1)),
    layers.Dense(5, activation= tf.keras.layers.LeakyReLU(alpha=0.1)),
    layers.Dense(num_classes)
])

model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
epochs = 50
history = model.fit(
    X_train, y_train,
    epochs=epochs, 
    validation_data = (X_val, y_val), 
    batch_size = 1000
)
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss=history.history['loss']
val_loss=history.history['val_loss']

epochs_range = range(epochs)

plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

```python
probability_model = tf.keras.Sequential([model, 
                                         tf.keras.layers.Softmax()])

prob = probability_model.predict(X_train)
print('Training AUC ROC with image:', roc_auc_score(y_train, prob[:, 1], multi_class='ovo'))

prob = probability_model.predict(X_val)
print('Validation AUC ROC with image:', roc_auc_score(y_val, prob[: , 1], multi_class='ovo'))


confusion = confusion_matrix(
y_val, prob.argmax(axis = 1))
confusion_normalized = confusion.astype("float") / confusion.sum(axis=1)

axis_labels = ['Non-hateful', 'Hateful']
ax = sns.heatmap(
    confusion, xticklabels=axis_labels, yticklabels=axis_labels,
    cmap='Blues', annot=True, fmt='.3f', square=True)
plt.title("Confusion matrix")
plt.ylabel("True label")
plt.xlabel("Predicted label")
plt.show()
```
