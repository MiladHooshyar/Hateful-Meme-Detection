```python
import numpy as np
import matplotlib.pylab as plt
from sklearn.manifold import TSNE
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, confusion_matrix
import seaborn as sns
```

```python
train = np.load('D:/meme_data/features/train.npz')
img_feature = train['img_fea']
text_feature= train['txt_fea']
label= train['label']


dev = np.load('D:/meme_data/features/dev.npz')
img_feature_dev = dev['img_fea']
text_feature_dev = dev['txt_fea']
label_dev= dev['label']
```

# TSNE of image features

```python
plt.figure(figsize=(5, 5))

plt.title('Image features')
X = img_feature[label==0, :]
X_embedded = TSNE(n_components=2).fit_transform(X)
plt.scatter(X_embedded[:, 0], X_embedded[:, 1], alpha = 0.1, c = 'r',
            s = 50, edgecolors=None, label = 'Non-hateful')

X = img_feature[label==1, :]
X_embedded = TSNE(n_components=2).fit_transform(X)
plt.scatter(X_embedded[:, 0], X_embedded[:, 1], alpha = 0.1, c = 'k',
            s = 50, edgecolors=None, label = 'Hateful')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

# TSNE of text features

```python
plt.figure(figsize=(5, 5))

plt.title('Text features')
X = text_feature[label==0, :]
X_embedded = TSNE(n_components=2).fit_transform(X)
plt.scatter(X_embedded[:, 0], X_embedded[:, 1], alpha = 0.1, c = 'r',
            s = 50, edgecolors=None, label = 'Non-hateful')

X = text_feature[label==1, :]
X_embedded = TSNE(n_components=2).fit_transform(X)
plt.scatter(X_embedded[:, 0], X_embedded[:, 1], alpha = 0.1, c = 'k',
            s = 50, edgecolors=None, label = 'Hateful')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

# TSNE of image+text features

```python
plt.figure(figsize=(5, 5))

plt.title('Image+Text features')
X = np.concatenate((img_feature[label==0, :], text_feature[label==0, :]), axis=1)
X_embedded = TSNE(n_components=2).fit_transform(X)
plt.scatter(X_embedded[:, 0], X_embedded[:, 1], alpha = 0.1, c = 'r',
            s = 50, edgecolors=None, label = 'Non-hateful')

X = np.concatenate((img_feature[label==1, :], text_feature[label==1, :]), axis=1)
X_embedded = TSNE(n_components=2).fit_transform(X)
plt.scatter(X_embedded[:, 0], X_embedded[:, 1], alpha = 0.1, c = 'k',
            s = 50, edgecolors=None, label = 'Hateful')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

# Baseline: Logistic Regression
## Image feature

```python

clf = LogisticRegression(random_state=0).fit(img_feature, label)
prob = clf.predict_proba(img_feature)
print('Training AUC ROC with image:', roc_auc_score(label, prob[:,1]))

prob = clf.predict_proba(img_feature_dev)
print('Validation AUC ROC with image:', roc_auc_score(label_dev, prob[:,1]))

print('******************')

confusion = confusion_matrix(
label_dev, prob.argmax(axis = 1))
confusion_normalized = confusion.astype("float") / confusion.sum(axis=1)

axis_labels = ['Non-hateful', 'Hateful']
ax = sns.heatmap(
    confusion_normalized, xticklabels=axis_labels, yticklabels=axis_labels,
    cmap='Blues', annot=True, fmt='.2f', square=True)
plt.title("Confusion matrix")
plt.ylabel("True label")
plt.xlabel("Predicted label")
plt.show()
```

# Baseline: Logistic Regression
## Image+Text feature

```python
X = np.concatenate((img_feature, text_feature), axis=1)
clf = LogisticRegression(random_state=0,max_iter = 1000).fit(X, label)
prob = clf.predict_proba(X)
print('Training AUC ROC with image:', roc_auc_score(label, prob[:,1]))

X = np.concatenate((img_feature_dev, text_feature_dev), axis=1)
prob = clf.predict_proba(X)
print('Validation AUC ROC with image:', roc_auc_score(label_dev, prob[:,1]))

print('******************')
confusion = confusion_matrix(
label_dev, prob.argmax(axis = 1))
confusion_normalized = confusion.astype("float") / confusion.sum(axis=1)

axis_labels = ['Non-hateful', 'Hateful']
ax = sns.heatmap(
    confusion_normalized, xticklabels=axis_labels, yticklabels=axis_labels,
    cmap='Blues', annot=True, fmt='.2f', square=True)
plt.title("Confusion matrix")
plt.ylabel("True label")
plt.xlabel("Predicted label")
plt.show()
```

# Baseline: Logistic Regression
## Image+Text feature
## Normliziation
## Validation split

```python
train = np.load('D:/meme_data/features/train.npz')
img_feature = train['img_fea']
text_feature= train['txt_fea']
label= train['label']


dev = np.load('D:/meme_data/features/dev.npz')
img_feature_dev = dev['img_fea']
text_feature_dev = dev['txt_fea']
label_dev= dev['label']
```

```python
val_ratio = 0.2
L_val = int(val_ratio * len(label))

label_train = label[:-L_val]
label_val = label[-L_val:]

text_feature_train = text_feature[:-L_val, :]
text_feature_val = text_feature[-L_val:, :]

img_feature_train = img_feature[:-L_val, :]
img_feature_val = img_feature[-L_val:, :]
```

```python
X1 = (img_feature_train - img_feature_train.mean(axis = 0)) / img_feature_train.std(axis = 0)
X2 = (text_feature_train - text_feature_train.mean(axis = 0)) / text_feature_train.std(axis = 0)
X = np.concatenate((X1, X2), axis=1)

clf = LogisticRegression(random_state=0,max_iter = 1000).fit(X, label_train)
prob = clf.predict_proba(X)
print('Training AUC ROC with image:', roc_auc_score(label_train, prob[:,1]))


X1 = (img_feature_val - img_feature_train.mean(axis = 0)) / img_feature_train.std(axis = 0)
X2 = (text_feature_val - text_feature_train.mean(axis = 0)) / text_feature_train.std(axis = 0)
X = np.concatenate((X1, X2), axis=1)
prob = clf.predict_proba(X)
print('Validation AUC ROC with image:', roc_auc_score(label_val, prob[:,1]))

print('******************')
confusion = confusion_matrix(
label_val, prob.argmax(axis = 1))
confusion_normalized = confusion.astype("float") / confusion.sum(axis=1)

axis_labels = ['Non-hateful', 'Hateful']
ax = sns.heatmap(
    confusion_normalized, xticklabels=axis_labels, yticklabels=axis_labels,
    cmap='Blues', annot=True, fmt='.2f', square=True)
plt.title("Confusion matrix")
plt.ylabel("True label")
plt.xlabel("Predicted label")
plt.show()
```
